```python
# z matrix ? 
CELL_N = len(df.irow(0).assign)
z_mat = np.zeros((CELL_N, CELL_N))
for r_i, r in df.iterrows():
    ca = r.assign
    for u in np.unique(ca):
        ent_i = np.argwhere(ca == u).flatten()
        for ci in ent_i:
            for cj in ent_i:
                z_mat[ci, cj] += 1

```

```python
import scipy.cluster.hierarchy as hier
from pandas import Categorical

lm = hier.linkage(z_mat, method='complete') # single complete average weighted ward centroid median
ord = np.array(hier.leaves_list(lm))

tgt_classes = []

true_assign_role = df.irow(0).true_assign_role
for ti, v in enumerate(true_assign_role):
    if v == None:
        true_assign_role[ti] = 'I'
        
ta = Categorical(true_assign_role)
ta_i = ta.labels
print ta.levels
fig = pylab.figure(figsize=(20, 12))

title_size = 20
tgt_cell_h = 39 # AS01
tgt_cell_v = 38 # AS06 

gs = gridspec.GridSpec(1, 3, width_ratios=[ 3.9, 1.5, 1]) 

gs.update(wspace=0.01, hspace=0.01)
axz = pylab.subplot(gs[0])

axz.imshow((z_mat[ord])[:, ord], interpolation='nearest', 
          cmap=pylab.cm.Greys)
axz.grid(False)
axz.set_xticks([])
axz.set_yticks([])
axz.set_title("cell / cell coassignment probability",fontsize= title_size)
axz.set_xlabel("cell ID", fontsize=title_size)
axz.set_ylabel("cell ID", fontsize=title_size)
axz.axhline(tgt_cell_h, c='r', alpha=0.5)
axz.axvline(tgt_cell_v, c='b', alpha=0.5)

axrole = pylab.subplot(gs[1])
neurons.cell_class.value_counts()


class_labels = []
class_pos = []
t_x = 0
role_colors = {'S' : 'y', 'M' : 'b', 'I' : 'g'}
for r_i, (role, pos) in enumerate([('M', 70), ('S', 10), ('I', 130)]):
    print r_i, role
    
    c_sorted = neurons[neurons.role == role].cell_class.value_counts().sort(inplace=False, ascending=False)[:10]

    classes = np.sort(c_sorted.index.values)
    tgt_classes.append(classes)
    

    for tci, tgt_class in enumerate(classes):
        axrole.axvline(t_x, c='k', alpha=0.1)
        r = np.array((neurons.cell_class == tgt_class))
        where = np.argwhere(r[ord]).flatten()
        axrole.scatter(np.ones(len(where))*t_x, where, c=role_colors[role], edgecolor='none', s=40)

        axrole.text(t_x, pos + 15*(tci % 2), tgt_class, ha='center', va = 'center', fontsize=12, rotation=90)
        t_x += 1.2
        class_pos.append(t_x-1.2)
        class_labels.append(tgt_class)
    t_x += 1.2
    
_ = axrole.set_xticks(class_pos)
_ = axrole.set_xticklabels([]) # class_labels, rotation='vertical', fontsize=8)
_ = axrole.set_yticks([])
axrole.grid(False)
axrole.set_ylim(279, 0)
axrole.set_title("anatomist cell class", fontsize=title_size)
#axrole.set_ylabel("cell ID", fontsize=title_size)

axrole.axhline(tgt_cell_h, c='r', alpha=0.5)
axsomapos = pylab.subplot(gs[2])
axsomapos.scatter(neurons.soma_pos[ord], np.arange(len(ord)), edgecolor='none', c='k')
axsomapos.grid(False)
axsomapos.set_xticks([0, 1.0])
axsomapos.set_yticks([])
axsomapos.set_ylim(279, 0)
axsomapos.set_title("soma position", fontsize=title_size)
axsomapos.axhline(tgt_cell_h, c='r', alpha=0.5)

fig.savefig("class_z_pos.pdf")
```

```python
dists = np.array([0.01, 0.1, 0.2, 0.5, 0.9])

def scale_conn(m):
    a = np.sort(m.flatten())
    b = a[len(a) * 0.999]
    # max is 80th percentile
    return np.minimum(m / b, 1.0)
    
mats = np.zeros((len(dists), 2, len(np.unique(ca)), len(np.unique(ca))))
for rel_i, rel in enumerate(['R1', 'R2']):

    latent_relation = latent['relations'][rel] 

    for di, d in enumerate(dists):
        e = irm.irmio.latent_distance_eval(d, latent_relation['ss'], latent_relation['hps'], "LogisticDistancePoisson" )
        # convert to canonical 
        for (g1, g2), p in e.iteritems():
            c1 = mapping[g1]
            c2 = mapping[g2]
            mats[di, rel_i, c1, c2] = p
```

```python

group_colors_opaque = group_colors.copy()
group_colors_opaque[:, -1] = 1.0
fig = pylab.figure(figsize=(20, 6))
ax = fig.add_subplot(1, 1, 1, aspect='equal') 
X_SCALE = 7.0
CIRCLE_SIZE = 0.4
for i in range(len(mats)):
    points = []
    colors = []
    widths = []
    
    patches = []

    for rel_i, color_i in enumerate([2, 0]):
        m = mats[i, rel_i]
        
        for ci in range(m.shape[0]):
            for cj in range(m.shape[1]):
                if ci != cj:
                    p = m[ci, cj] / 20.0
                    if p >= 0.2:
                        points.append([[i*X_SCALE, ci], [(i+1)*X_SCALE, cj]])
                        color = np.zeros(4)
                        color[-1] = min(p, 1.0)
                        color[color_i] = 1.0
                        colors.append(color)
                        widths.append((1+p))

                    
    for ci in range(mats.shape[2]):
        circle = Circle((i*X_SCALE,ci), CIRCLE_SIZE)
        patches.append(circle)
    
    lc = LineCollection(points, colors=colors, linewidths=np.array(widths), zorder=1)
    ax.add_collection(lc)

    p = PatchCollection(patches, color = group_colors_opaque, edgecolor='k', zorder=10) # , color='w', edgecolor='k')

    ax.add_collection(p)
# last ones
patches = [Circle((mats.shape[0]*X_SCALE,ci), CIRCLE_SIZE) for ci in range(mats.shape[2])]
p = PatchCollection(patches, color = group_colors_opaque, edgecolor='k', zorder=10)
ax.add_collection(p)

ax.set_yticks([])
ax.set_xticks(np.arange(mats.shape[0] + 1) * X_SCALE)
ax.set_xticklabels([""] + ["%2.0f %%"% (x*100) for x in  dists], fontsize=20)
ax.set_xlim(-1, len(mats)*X_SCALE + 1)
ax.set_ylim(mats.shape[2], -1)
ax.set_xlabel("cell distance (frac of body length)", fontsize=20)
ax.set_ylabel("discovered cell type", fontsize=20)

text_start_pos = len(mats)*X_SCALE + 1
for g_i, g in neurons.groupby('sample_sorted'):
    
    vc = g.cell_class.value_counts()
    vc_names = vc.index.values
    vc_values = np.array(vc)
    group_example_str = ""
    for i in range(min(3, len(vc_values))):
        group_example_str += "%s (%d) " % (vc_names[i], vc_values[i])

    ax.text(text_start_pos, g_i, group_example_str, verticalalignment='center', fontsize=12)
    
fig.savefig("tgt.bestsample.connections.pdf")
```
