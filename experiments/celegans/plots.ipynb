```python
# z matrix ? 
CELL_N = len(df.irow(0).assign)
z_mat = np.zeros((CELL_N, CELL_N))
for r_i, r in df.iterrows():
    ca = r.assign
    for u in np.unique(ca):
        ent_i = np.argwhere(ca == u).flatten()
        for ci in ent_i:
            for cj in ent_i:
                z_mat[ci, cj] += 1

```

```python
import scipy.cluster.hierarchy as hier
from pandas import Categorical

lm = hier.linkage(z_mat, method='complete') # single complete average weighted ward centroid median
ord = np.array(hier.leaves_list(lm))

tgt_classes = []

true_assign_role = df.irow(0).true_assign_role
for ti, v in enumerate(true_assign_role):
    if v == None:
        true_assign_role[ti] = 'I'
        
ta = Categorical(true_assign_role)
ta_i = ta.labels
print ta.levels
fig = pylab.figure(figsize=(20, 12))

title_size = 20
tgt_cell_h = 39 # AS01
tgt_cell_v = 38 # AS06 

gs = gridspec.GridSpec(1, 3, width_ratios=[ 3.9, 1.5, 1]) 

gs.update(wspace=0.01, hspace=0.01)
axz = pylab.subplot(gs[0])

axz.imshow((z_mat[ord])[:, ord], interpolation='nearest', 
          cmap=pylab.cm.Greys)
axz.grid(False)
axz.set_xticks([])
axz.set_yticks([])
axz.set_title("cell / cell coassignment probability",fontsize= title_size)
axz.set_xlabel("cell ID", fontsize=title_size)
axz.set_ylabel("cell ID", fontsize=title_size)
axz.axhline(tgt_cell_h, c='r', alpha=0.5)
axz.axvline(tgt_cell_v, c='b', alpha=0.5)

axrole = pylab.subplot(gs[1])
neurons.cell_class.value_counts()


class_labels = []
class_pos = []
t_x = 0
role_colors = {'S' : 'y', 'M' : 'b', 'I' : 'g'}
for r_i, (role, pos) in enumerate([('M', 70), ('S', 10), ('I', 130)]):
    print r_i, role
    
    c_sorted = neurons[neurons.role == role].cell_class.value_counts().sort(inplace=False, ascending=False)[:10]

    classes = np.sort(c_sorted.index.values)
    tgt_classes.append(classes)
    

    for tci, tgt_class in enumerate(classes):
        axrole.axvline(t_x, c='k', alpha=0.1)
        r = np.array((neurons.cell_class == tgt_class))
        where = np.argwhere(r[ord]).flatten()
        axrole.scatter(np.ones(len(where))*t_x, where, c=role_colors[role], edgecolor='none', s=40)

        axrole.text(t_x, pos + 15*(tci % 2), tgt_class, ha='center', va = 'center', fontsize=12, rotation=90)
        t_x += 1.2
        class_pos.append(t_x-1.2)
        class_labels.append(tgt_class)
    t_x += 1.2
    
_ = axrole.set_xticks(class_pos)
_ = axrole.set_xticklabels([]) # class_labels, rotation='vertical', fontsize=8)
_ = axrole.set_yticks([])
axrole.grid(False)
axrole.set_ylim(279, 0)
axrole.set_title("anatomist cell class", fontsize=title_size)
#axrole.set_ylabel("cell ID", fontsize=title_size)

axrole.axhline(tgt_cell_h, c='r', alpha=0.5)
axsomapos = pylab.subplot(gs[2])
axsomapos.scatter(neurons.soma_pos[ord], np.arange(len(ord)), edgecolor='none', c='k')
axsomapos.grid(False)
axsomapos.set_xticks([0, 1.0])
axsomapos.set_yticks([])
axsomapos.set_ylim(279, 0)
axsomapos.set_title("soma position", fontsize=title_size)
axsomapos.axhline(tgt_cell_h, c='r', alpha=0.5)

fig.savefig("class_z_pos.pdf")
```
