```python
# z matrix ? 
CELL_N = len(df.irow(0).assign)
z_mat = np.zeros((CELL_N, CELL_N))
for r_i, r in df.iterrows():
    ca = r.assign
    for u in np.unique(ca):
        ent_i = np.argwhere(ca == u).flatten()
        for ci in ent_i:
            for cj in ent_i:
                z_mat[ci, cj] += 1

```

```python
import scipy.cluster.hierarchy as hier
from pandas import Categorical

lm = hier.linkage(z_mat, method='single') # single complete average weighted ward centroid median
ord = np.array(hier.leaves_list(lm))

tgt_classes = []

#true_assign_role = df.irow(0).true_assign_role
#for ti, v in enumerate(true_assign_role):
#    if v == None:
#        true_assign_role[ti] = 'I'
        
#ta = Categorical(true_assign_role)
#ta_i = ta.labels
#print ta.levels
fig = pylab.figure(figsize=(20, 12))

title_size = 20
tgt_cell_h = 39 # AS01
tgt_cell_v = 38 # AS06 

gs = gridspec.GridSpec(1, 3, width_ratios=[ 5.3, 1.5, 1.0 + 1.0]) 

gs.update(wspace=0.01, hspace=0.01)
axz = pylab.subplot(gs[0])
axrole = pylab.subplot(gs[1])
axprofile = pylab.subplot(gs[2])

axz.imshow((z_mat[ord])[:, ord], interpolation='nearest', 
          cmap=pylab.cm.Greys)
axz.grid(False)
axz.set_xticks([])
axz.set_yticks([])
axz.set_title("cell / cell coassignment probability",fontsize= title_size)
axz.set_xlabel("cell ID", fontsize=title_size)
axz.set_ylabel("cell ID", fontsize=title_size)
#axz.axhline(tgt_cell_h, c='r', alpha=0.5)
#axz.axvline(tgt_cell_v, c='b', alpha=0.5)



neurons['coarse'][neurons['type_id'] == 58] = 'bc'
cmap = {None : 4, 
     'other': 4, 
     'nac' : 2, 
     'bc' : 1, 
     'mwac' : 3, 
     'gc' : 0}
color_idx = [cmap[o] for o in neurons['coarse']]        
import brewer2mpl
cmap = brewer2mpl.get_map('Set1', 'qualitative', 5).mpl_colormap
#     typeax.scatter(cell_types, np.argsort(ca), edgecolor='none', 
#                s=3, 
#                c=color_idx, 
#                alpha=0.8,
#                cmap = cmap)
p = np.array([0, 12, 24, 57, 72, 80])
axrole.set_xticks(p)
axrole.set_xticks(p[1:] - np.diff(p)/2. , minor=True)
axrole.set_yticks([])
axrole.set_xticklabels(['gc', 'bc', 'nac', 'mwac', 'other'], minor=True,
                               fontsize=14, rotation=90)


axrole.scatter(neurons.type_id, np.argsort(ord), 
               edgecolor='none', c=color_idx, s=10,
               cmap=cmap)
[axrole.axvline(x, linestyle='--', c='k', alpha=0.5,  zorder=-100) for x in p]

axrole.grid(False)
axrole.set_ylim(len(ord), 0)
axrole.set_title("anatomist cell class", fontsize=title_size)
axrole.set_xticklabels([])


bins = np.linspace(30, 125, 50)
synimg = np.zeros((len(neurons), len(bins)-1))

for i, contacts in enumerate(neurons.contact_x_list):
    if isinstance(contacts, list):
        
        synimg[i], _ = np.histogram(contacts, bins)
        
        
extent=[30, 125, len(ord), 0] 

        
axprofile.imshow(synimg[ord], extent=extent, aspect=0.3, cmap=pylab.cm.Blues)  #set_ylim(len(ord), 0)
axprofile.set_yticks([])
axprofile.grid(False)
axprofile.set_xlabel("um")
axprofile.set_title("contact density vs depth")

axprofile.scatter(neurons.x, np.argsort(ord), edgecolor='none', s=10, c='k', alpha=0.7)
axprofile.set_ylim(len(ord), 0)
#axprofile.set_xlim(40, 110)
#axprofile.set_xticks([40, 110])
#axprofile.grid(False)
#axprofile.set_xlabel("um")
#axprofile.set_title("soma depth")


#axsomadepth.grid(False)
#axsomapos.set_xticks([0, 1.0])
#axsomapos.set_yticks([])
#axsomapos.set_ylim(279, 0)
#axsomapos.set_title("soma position", fontsize=title_size)
#axsomapos.axhline(tgt_cell_h, c='r', alpha=0.5)

fig.savefig("tgt.class_z_pos_depth.pdf")
```
